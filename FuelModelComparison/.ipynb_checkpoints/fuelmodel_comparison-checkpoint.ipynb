```python
#2. Trip Level 
def mpg_legs(legs):
    legs['mpg'] = (legs['total distance (m)']*0.000621371) / (legs['total fuel consumed']*0.264172)
    return  legs
```

```python
map_legs = mpg_legs(map_legs)
routee_legs = mpg_legs(routee_legs)
```

```python
map_median = int(map_legs['mpg'].median())
routee_median = int(routee_legs['mpg'].median())
```

```python
plt.figure(figsize = (10,5))
plt.hist(map_legs['mpg'],bins = 50, density=True, label = 'Map', facecolor='g', alpha=0.75)
plt.axvline(map_median,label = 'Median: {} mpg'.format(map_median), color='g')

plt.hist(routee_legs['mpg'],bins = 50, density=True, label = 'RouteE', facecolor='b', alpha=0.75)
plt.axvline(routee_median, label = 'Median: {} mpg'.format(routee_median),color='b' )

plt.title("Distribution of Trip Mileage for Fuel Models")
plt.xlabel("Miles per Gallon")

plt.legend();
```

```python
#3. Segmentation by Trip Distance - Short (<5 miles), Medium (5-10 miles), Long (>10 miles)
```

```python
plt.hist(map_legs['total distance (m)']*0.000621371, bins = 50)
plt.title("Distribution of Trip distance")
plt.xlabel("Miles");
```

```python
map_legs['total distance (m)'].describe()*0.000621371
```

```python
def tripdistance_segmentation(df = map_legs, col = 'total distance (m)'):
    conditions =  [ df[col] <= 5*1609.34 , (df[col] <= 10*1609.34) & (df[col]> 5 *1609.34), df[col] > 10*1609.34 ]
    choices = [ "short", 'medium', 'long' ]
    df['tripdistance_segmentation'] = np.select(conditions, choices, default=np.nan)
    
def plot_histogram(f1 = map_legs, f2 = routee_legs, seg = ['short'], title = ""):

    df1 = f1[f1['tripdistance_segmentation'].isin(seg)]
    df2 = f2[f2['tripdistance_segmentation'].isin(seg)]
    map_median = int(df1['mpg'].median())
    routee_median = int(df2['mpg'].median())

    plt.figure(figsize = (10,5))
    plt.hist(df1['mpg'],bins = 50, density=True, label = 'Map', facecolor='g', alpha=0.75)
    plt.axvline(map_median,label = 'Median: {} mpg'.format(map_median), color='g')

    plt.hist(df2['mpg'],bins = 50, density=True, label = 'RouteE', facecolor='b', alpha=0.75)
    plt.axvline(routee_median, label = 'Median: {} mpg'.format(routee_median),color='b' )
    
    plt.xlabel("Miles per Gallon")
    plt.title(title)
    plt.legend();
```

```python
tripdistance_segmentation(df = map_legs, col = 'total distance (m)')
tripdistance_segmentation(df = routee_legs, col = 'total distance (m)')
map_legs['tripdistance_segmentation'].value_counts()*100/len(map_legs)
```

```python
plot_histogram(f1 = map_legs, f2 = routee_legs, seg = ['short'], title = "Distribution of Trip Mileage for Fuel Models - Short Trips (<=5 miles)")
```

```python
plot_histogram(f1 = map_legs, f2 = routee_legs, seg = ['medium'], title = "Distribution of Trip Mileage for Fuel Models - Medium Trips (5-10 miles)")
```

```python
plot_histogram(f1 = map_legs, f2 = routee_legs, seg = ['long'], title = "Distribution of Trip Mileage for Fuel Models - Long Trips ( >10 miles)")
```

```python
## 4. Fuel Map model - mpg changes a lot with consumption
```

```python
map_legs['congestedtt'] = pd.to_datetime(map_legs['duration (congested)'],errors = 'coerce',format = "%H:%M:%S.%f")
map_legs['congested_ttsec'] = map_legs['congestedtt'].dt.hour*3600+ map_legs['congestedtt'].dt.minute*60 + map_legs['congestedtt'].dt.second
    
```

```python
map_legs['speed_mph'] = (map_legs['total distance (m)']*0.000621371)/(map_legs['congested_ttsec']*0.000277778)
map_legs.groupby('tripdistance_segmentation')['speed_mph'].median()
```

```python
map_legs['speed_mph'].describe() #
```

```python
#null trip speeds - why? 
dfsub = map_legs[(map_legs == np.inf).any(axis=1)]
dfsub['tripdistance_segmentation'].value_counts() # all very very short trips 
```

```python
len(map_legs[(map_legs['speed_mph'] != np.inf)]), len(map_legs) # remove them
```

```python
def plot_histogram_speed(f1 = map_legs, seg = ['short'], title = ""):

    df1 = f1[(f1['tripdistance_segmentation'].isin(seg)) & (f1['speed_mph'] != np.inf)] # seg and remove null
    map_median = int(df1['speed_mph'].median())

    plt.figure(figsize = (10,5))
    plt.hist(df1['speed_mph'],bins = 50, density=True, label = 'Map', facecolor='g', alpha=0.75)
    plt.axvline(map_median,label = 'Median: {} mph'.format(map_median), color='g')
    
    plt.xlabel("Trip speed (miles per hour)")
    plt.title(title)
    plt.legend();
```

```python
plot_histogram_speed(f1 = map_legs, seg = ['short'], title = "Distribution of Trip Speed (Fuel Map Model)- Short Trips (<=5 miles)")
```

```python
plot_histogram_speed(f1 = map_legs, seg = ['medium'], title = "Distribution of Trip Speed (Fuel Map Model)- Medium Trips (5-10 miles)")
```

```python
plot_histogram_speed(f1 = map_legs, seg = ['long'], title = "Distribution of Trip Speed (Fuel Map Model)- Long Trips (>10 miles)")
```
